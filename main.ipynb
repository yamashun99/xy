```python
import numpy as np
import matplotlib.pyplot as plt
import qmc
import importlib

importlib.reload(qmc)
from qmc import *
```

```python
J = 100
L = 10
M = 40
beta = 10
conf0 = np.ones((L, M)) * -1
conf0[1, :] = 1
c = Conf(J, L, M, beta)
c.conf = conf0
```

```python
fig, ax = plt.subplots()
for ix in range(L):
    for iy in range(M):
        if c.conf[ix][iy] == 1:
            ax.scatter(ix, iy, c="r")
        else:
            ax.scatter(ix, iy, c="b")
for i in range(L // 2):
    for j in range(M // 2):
        ax.fill_between([2 * i, 2 * i + 1], 2 * j, 2 * j + 1, facecolor="b", alpha=0.1)
        ax.fill_between(
            [2 * i + 1, 2 * i + 2], 2 * j + 1, 2 * j + 2, facecolor="b", alpha=0.1
        )
```
